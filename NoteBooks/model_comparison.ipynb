```python
import matplotlib.pyplot as plt

models = ["CNN", "EfficientNet"]
accuracy = [94.35, 83.72]

plt.figure()
plt.bar(models, accuracy)
plt.title("Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy (%)")

for i, v in enumerate(accuracy):
    plt.text(i, v + 0.5, f"{v:.2f}", ha='center')

plt.show()
```
